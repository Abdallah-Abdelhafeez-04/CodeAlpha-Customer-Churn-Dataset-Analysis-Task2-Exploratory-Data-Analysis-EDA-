```python
import pandas as pd
```

```python
df=pd.read_csv('churn-bigml-20.csv')
df
```

```python
# Basic structure
df.info()

# Summary statistics
df.describe(include="all")
```

```python
# Overall churn rate
df['Churn_flag'] = df['Churn'].astype(int)
print("Churn distribution:")
print(df['Churn_flag'].value_counts(normalize=True).round(3))
```

```python
print("Churn rate by International plan:")
print(df.groupby('International plan')['Churn_flag'].mean().round(3))
```

```python
print("Churn rate by Voice mail plan:")
print(df.groupby('Voice mail plan')['Churn_flag'].mean().round(3))
```

```python
print("Churn rate by number of customer service calls:")
print(df.groupby('Customer service calls')['Churn_flag'].mean().round(3))
```

```python
# Look for extreme values in day/eve/night minutes
for col in ['Total day minutes','Total eve minutes','Total night minutes','Total intl minutes']:
    print(f"\n{col} outliers (top 5):")
    print(df[[col,'Churn_flag']].sort_values(by=col, ascending=False).head())
```

Expectation: ~14% churn rate.

```python
import matplotlib.pyplot as plt
import seaborn as sns
# International plan
print(df.groupby('International plan')['Churn_flag'].mean())
sns.barplot(x='International plan', y='Churn_flag', data=df)
plt.title('Churn rate by International Plan')
plt.show()

# Voice mail plan
print(df.groupby('Voice mail plan')['Churn_flag'].mean())
sns.barplot(x='Voice mail plan', y='Churn_flag', data=df)
plt.title('Churn rate by Voice Mail Plan')
plt.show()

# Area code (top 5 only for clarity)
top_areas = df['Area code'].value_counts().nlargest(5).index
print(df[df['Area code'].isin(top_areas)].groupby('Area code')['Churn_flag'].mean())
sns.barplot(x='Area code', y='Churn_flag', data=df[df['Area code'].isin(top_areas)])
plt.title('Churn rate by Top 5 Area Codes')
plt.show()
```

Expectation:

International plan customers churn much more (~36% vs

12%).

No voicemail plan = higher churn.

Area code differences exist but less strong.

```python
# Summary statistics
print(df.describe())

# Compare distributions of call minutes by churn
num_cols = ['Total day minutes','Total eve minutes','Total night minutes','Total intl minutes','Customer service calls']

for col in num_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(x='Churn_flag', y=col, data=df)
    plt.title(f'{col} by Churn')
    plt.show()
```

Expectation: churners often have higher customer service calls and more international minutes.

```python
df.groupby('Churn')['Customer service calls'].mean()
```

If churners average more calls → evidence supports hypothesis.

```python
df.groupby('Customer service calls')['Churn'].mean().sort_index()
```

This gives churn probability by number of calls (e.g., 0 calls → 5% churn, 4+ calls → 40% churn).

```python
df['High_service_calls'] = (df['Customer service calls'] >= 4).astype(int)
df.groupby('High_service_calls')['Churn'].mean()
```

Hypothesis: customers with ≥4 service calls have much higher churn probability.

Churners ≈ 14% overall.

But among customers with ≥4 service calls, churn rate can jump to ~40–45%.

Customers with 0–1 calls have very low churn (≈5–7%).

This supports the hypothesis that customer service calls are a strong indicator of dissatisfaction → higher churn risk.

```python


# Create churn flag (0/1)
df['Churn_flag'] = df['Churn'].astype(int)

# Plot churn rate by number of service calls
plt.figure(figsize=(7,5))
sns.barplot(
    x="Customer service calls",
    y="Churn_flag",
    data=df,
    estimator=lambda x: sum(x)/len(x),  # churn rate
    ci=None,
    palette="viridis"
)

plt.ylabel("Churn Rate")
plt.title("Churn Rate by Number of Customer Service Calls")
plt.show()
```
